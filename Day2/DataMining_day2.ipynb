# Lab Assignment: Recapitulation of Python, Numpy, Pandas: Rule-based Classifier

```python
import pandas as pd

# Exact data from the image table
data = {
    'Name': [
        'human', 'python', 'salmon', 'whale', 'frog',
        'komodo', 'bat', 'pigeon', 'cat', 'leopard shark',
        'turtle', 'penguin', 'porcupine', 'eel', 'salamander',
        'gila monster', 'platypus', 'owl', 'dolphin', 'eagle'
    ],
    'Blood Type': [
        'warm', 'cold', 'cold', 'warm', 'cold',
        'cold', 'warm', 'warm', 'warm', 'cold',
        'cold', 'warm', 'warm', 'cold', 'cold',
        'cold', 'warm', 'warm', 'warm', 'warm'
    ],
    'Give Birth': [
        'yes', 'no', 'no', 'yes', 'no',
        'no', 'yes', 'no', 'yes', 'yes',
        'no', 'no', 'yes', 'no', 'no',
        'no', 'no', 'no', 'yes', 'no'
    ],
    'Can Fly': [
        'no', 'no', 'no', 'no', 'no',
        'no', 'yes', 'yes', 'no', 'no',
        'no', 'no', 'no', 'no', 'no',
        'no', 'no', 'yes', 'no', 'yes'
    ],
    'Live in Water': [
        'no', 'no', 'yes', 'yes', 'sometimes',
        'no', 'no', 'no', 'no', 'yes',
        'sometimes', 'no', 'no', 'yes', 'sometimes',
        'sometimes', 'no', 'no', 'yes', 'no'
    ]
}

df = pd.DataFrame(data)
```

```python

df.to_csv('vertebrate_data.csv', index=False)


print("===== Vertebrate Data =====")
print(df)
```

```python
print("\nShape of the dataset (rows, columns):", df.shape)

print("\n===== Unique Values in Each Column =====")
for column in df.columns:
    print(f"\n{column}:\n{df[column].value_counts()}")
```

```python
# Rule-based classifier function
def classify(row):
    if row['Give Birth'] == 'no' and row['Can Fly'] == 'yes':
        return 'Birds'           # R1
    elif row['Give Birth'] == 'no' and row['Live in Water'] == 'yes':
        return 'Fishes'          # R2
    elif row['Give Birth'] == 'yes' and row['Blood Type'] == 'warm':
        return 'Mammals'         # R3
    elif row['Give Birth'] == 'no' and row['Can Fly'] == 'no':
        return 'Reptiles'        # R4
    elif row['Live in Water'] == 'sometimes':
        return 'Amphibians'      # R5
    else:
        return 'Unknown'
```

```python
# Apply the classifier
df['Class'] = df.apply(classify, axis=1)

# Display final classified dataset
print("\n=====  Vertebrate Dataset with Predicted Class =====")
print(df)
```
